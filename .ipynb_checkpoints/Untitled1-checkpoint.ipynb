```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
df = pd.read_csv('personality_scores.csv', sep=";")
df.head()
```

```python
def assigning(row,Category_Num):
    count = 0
    asd = 0
    for column in df.columns:
        tuplee = row[column]
        if type(tuplee) == str:
            catergory = tuplee[1]
            result = int(tuplee[4])
            if catergory == Category_Num:
                count += result
    return count
```

```python
df['Extraversion'] = df.apply(lambda row : assigning(row,'1'), axis = 1)
df['Agreeableness'] = df.apply(lambda row : assigning(row,'2'), axis = 1)
df['Conscientiousness'] = df.apply(lambda row : assigning(row,'3'), axis = 1)
df['Emotional Stability'] = df.apply(lambda row : assigning(row,'4'), axis = 1)
df['Openness to experiences'] = df.apply(lambda row : assigning(row,'5'), axis = 1)
```
