### 2. Create new columns containing the total score of each of the personality test subscales.

```python
# funtion for creating the scores of the new columns 
def convertData(row, Nums_Cat):
    count = 0
    for values in df.columns:
        lst = row[values]
        if type(lst) == str:
            cat = lst[1]
            result = int(lst[4])
            if cat == Nums_Cat:
                count += result
    return count 
            
```

```python
# applying the function above to the new columns 
df['Extraversion'] = df.apply(lambda row : convertData(row,'1'), axis = 1)
df['Agreeableness'] = df.apply(lambda row : convertData(row,'2'), axis = 1)
df['Conscientiousness'] = df.apply(lambda row : convertData(row,'3'), axis = 1)
df['Emotional Stability'] = df.apply(lambda row : convertData(row,'4'), axis = 1)
df['openness'] = df.apply(lambda row : convertData(row,'5'), axis = 1)
```

```python
# printing the new dataset with new columns
df.head()
```

#### 2. Creating a condition that has to be met by the columns (named) and a new DataFrames 

```python
# finding the values that are less than 30 in the three columns 
df_merge[(df_merge['Agreeableness'] < 30) & (df_merge['Emotional Stability'] < 30) & (df_merge['Conscientiousness'] < 30)]                      
                        
```

```python
# creating a new DataFrame with the columns stated below
Risk_status = pd.DataFrame(df_merge[['ID','Extraversion','Agreeableness','Conscientiousness','Emotional Stability','openness','Department']])
Risk_status.head()
```

```python
# funtion that is creating the high and low of the new columns
def risk_status(score, name):
    mean = df_merge[name].mean()
    standiv = df_merge[name].std()
    high = (mean + standiv)/2
    low = (mean - standiv)/2
    if score >= high:
        return 'high'
    else:
        return 'low'
```

```python
# creating new columns for the merged DataFrame and calling the relevant columns  
df_merge['Extraversion levels'] = df_merge['Extraversion'].apply(lambda column : risk_status(column,'Extraversion'))
df_merge['Agreeableness levels'] = df_merge['Agreeableness'].apply(lambda column : risk_status(column,'Agreeableness'))
df_merge['Conscientiousness levels'] = df_merge['Conscientiousness'].apply(lambda column : risk_status(column,'Conscientiousness'))
df_merge['Emotional Stability levels'] = df_merge['Emotional Stability'].apply(lambda column : risk_status(column,'Emotional Stability'))
df_merge['openness levels'] = df_merge['openness'].apply(lambda column : risk_status(column,'openness'))
df_merge.loc[:, 'Agreeableness':'openness levels']
```

```python
# we are creating a new column called  
Risk_status['Risk Status'] = np.where(
    (Risk_status['Emotional Stability'] < 30) & \
    (Risk_status.Conscientiousness < 30) & (Risk_status.Agreeableness < 30), 'High risk', 'Low risk')
Risk_status.head()
```

```python
# we are calling the columns, grouping them and making them into a DataFrame 
summary = Risk_status[['Risk Status', 'Department']].groupby(['Risk Status', 'Department']).size()
pd.DataFrame(summary)
```

```python
summary = pd.pivot_table(summary, index=['Risk Status'],  columns=['Department'], margins=True, aggfunc=np.sum) 
```
