### 2. Create new columns containing the total score of each of the personality test subscales.

```python
def convertData(row, Nums_Cat):
    count = 0
    touch = 0
    for values in df.columns:
        lst = row[values]
        if type(lst) == str:
            cat = lst[1]
            result = int(lst[4])
            if cat == Nums_Cat:
                count += result
       
    return count 
            
```

```python
df['Extraversion'] = df.apply(lambda row : convertData(row,'1'), axis = 1)
df['Openness'] = df.apply(lambda row : convertData(row,'2'), axis = 1)
df['Agreeable'] = df.apply(lambda row : convertData(row,'3'), axis = 1)
df['Emotional Stablility'] = df.apply(lambda row : convertData(row,'4'), axis = 1)
df['Conscienctiousness'] = df.apply(lambda row : convertData(row,'5'), axis = 1)
```
