### 2. Create new columns containing the total score of each of the personality test subscales.

```python
def convertData(row, Nums_Cat):
    count = 0
    touch = 0
    for values in df.columns:
        lst = row[values]
        if type(lst) == str:
            cat = lst[1]
            result = int(lst[4])
            if cat == Nums_Cat:
                count += result
       
    return count 
            
```

```python
df['Conscienctiousness'] = df.apply(lambda row : convertData(row,'1'), axis = 1)
df['Emotional Stablility'] = df.apply(lambda row : convertData(row,'2'), axis = 1)
df['Openness'] = df.apply(lambda row : convertData(row,'3'), axis = 1)
df['Agreeable'] = df.apply(lambda row : convertData(row,'4'), axis = 1)
df['Extraversion'] = df.apply(lambda row : convertData(row,'5'), axis = 1)
```

## 1. Importing the data and merging data

```python
df_departments = pd.read_csv('departments.csv', sep=';')
df_departments.head()
```

```python
df_merge = pd.merge(df, df_departments, on=None, how='outer')
df_merge.head()
```

```python
assert len(df_departments) == len(df_merge)
```

```python
df_merge[(df_merge['Agreeable'] < 30) & (df_merge['Emotional Stablility'] < 30) & (df_merge['Conscienctiousness'] < 30)]                      
                        
```

```python
def risk_status(score, name):
    mean = df_merge[name].mean()
    standiv = df_merge[name].std()
    high = (mean + standiv)/2
    low = (mean - standiv)/2
    if score >= high:
        return 'high'
    else:
        return 'low'
```

```python
df['Conscienctiousness levels'] = df_merge['Conscienctiousness'].apply(lambda column : risk_status(column,'Conscienctiousness'))
df['Emotional Stablility levels'] = df_merge['Emotional Stablility'].apply(lambda column : risk_status(column,'Emotional Stablility'))
df['Openness levels'] = df_merge['Openness'].apply(lambda column : risk_status(column,'Openness'))
df['Agreeable levels'] = df_merge['Agreeable'].apply(lambda column : risk_status(column,'Agreeable'))
df['Extraversion levels'] = df_merge['Extraversion'].apply(lambda column : risk_status(column,'Extraversion'))
```
